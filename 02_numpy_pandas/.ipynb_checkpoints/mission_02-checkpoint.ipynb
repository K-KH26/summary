### 1.(Python2_15) 아래 요구사항에 맞게 코드를 작성해주세요. (30점)
새로운 사용자의 email을 받으면 데이터프레임에 추가하는 함수를 만들려 합니다
- 함수는 데이터프레임과 email을 받고 데이터프레임을 반환합니다 \
(email은 영문자 소문자, 숫자로만 이루어져 있습니다)
- id는 항상 가장 마지막 id보다 1 큰 문자열입니다
- user_name은 email 주소의 id에 해당합니다
- domain은 email 컬럼에서 메일의 도메인만 가져옵니다
- 'user_name'과 'domain'은 split()으로도, re로도 가져올 수 있습니다

**결과 예시**
```python
# 함수 실행
df_update = update_user_data(df, 'fastcamp22@gmail.com')
print(df_update)
```
```
# 결과
         id                 email   user_name domain
0  62040702         kim@gmail.com         kim  gmail
1  62040703          ko@naver.com          ko  naver
2  62040704         joh@gmail.com         joh  gmail
3  62040705         lee@naver.com         lee  naver
4  62040706        park@gmail.com        park  gmail
5  62040707  fastcamp22@gmail.com  fastcamp22  gmail
```

````python
# import pandas as pd

data = {
    "id" : ['62040702','62040703','62040704','62040705','62040706'],
    "email" : ['kim@gmail.com','ko@naver.com','joh@gmail.com','lee@naver.com','park@gmail.com'],
    "user_name" : ['kim','ko','joh','lee','park'],
    "domain" : ['gmail','naver','gmail','naver','gmail']
}
df = pd.DataFrame(data)
df
````

````python
# 여기에 코드를 작성하세요
def update_user_data(df, email):
    # email 스플릿, 이름, 도메인
    name = email.split("@")[0]
    domain = email.split("@")[1].split(".")[0]
    
    # df의 마지막 인덱스 id => int형으로 변환 + 1 => 다시 str로 변환
#     id = df.loc[len(df) - 1]["id"]
#     id = df.index[-1]
    new_id = str( int(df.tail(1)["id"]) + 1 )
    
    # 데이터 추가.
    user_data = [new_id, email, name, domain]
    df.loc[len(df)] = user_data
    
    return df
````

````python
# 실행 코드
df_update = update_user_data(df, 'fastcamp22@gmail.com')
df_update
````

아래 입양 동물 데이터에서 개/고양이에 따른 입양 속도(AdoptionSpeed)를 비율로 보려합니다 
- 제공된 df을 받아 df_percent를 return하는 함수를 완성해주세요
- 개/고양이 기준 입양속도별 개수를 나타내는 새로운 df_percent를 만들어주세요
- 각 합을 나타내는 total column을 만들어주세요
- 각 숫자를 total 기준으로 퍼센트 값으로 바꿔주세요 
- 숫자를 세기 위해선 agg function으로 count를 쓰면 됩니다
- 개/고양이에 따른 입양 속도를 보려면 index론 type, column으론 adoptionspeed를 두어야합니다
- pandas에는 특정 row의 값들의 합을 구해주는 함수가 있습니다

**결과 예시** \
```
AdoptionSpeed	0	1	2	3	4	total
Type						
Cat	3.71	24.74	28.12	18.88	24.55	100.0
Dog	2.21	18.35	26.74	24.41	28.30	100.0
```

````python
# 문제 풀기 전에 실행해주세요
import pathlib
import tensorflow as tf
import pandas as pd

dataset_url = 'http://storage.googleapis.com/download.tensorflow.org/data/petfinder-mini.zip'
csv_file = 'datasets/petfinder-mini/petfinder-mini.csv'
tf.keras.utils.get_file('petfinder_mini.zip', dataset_url, extract=True, cache_dir='.')
df = pd.read_csv(csv_file)

df = df[['Type','Age','Gender','Gender','Fee', 'AdoptionSpeed']]
df
````

````python
# 여기에 코드를 작성하세요
def get_adoptspeed_percent(df):
    df["counts"] = 1
    
    df_percent = df.pivot_table("counts",["Type"],["AdoptionSpeed"], aggfunc=sum)
    df_percent["total"] = df_percent[0] + df_percent[1] + df_percent[2] + df_percent[3] + df_percent[4]
    
    cat_total = df_percent.loc["Cat"]["total"]
    dog_total = df_percent.loc["Dog"]["total"]
    
    cat_percent = round((df_percent.loc["Cat"]/cat_total) * 100, 2)
    dog_percent = round((df_percent.loc["Dog"]/dog_total) * 100, 2)
    
    df_percent.loc["Cat"] = cat_percent
    df_percent.loc["Dog"] = dog_percent
    
    return df_percent

df_percent = get_adoptspeed_percent(df)
df_percent
````

````python
# 문제 풀기 전에 실행해주세요
from sklearn.datasets import load_boston

dataset = load_boston()
raw_data = pd.DataFrame(dataset.data, columns=dataset.feature_names)
raw_data['MEDV'] = dataset.target
raw_data.head()
````

````python
# 선형회귀 모델
from sklearn import linear_model
# 학습 데이터와 테스트 데이터를 나눠주는 모듈
from sklearn.model_selection import train_test_split
# 모델을 평가해주는 모듈
from sklearn.metrics import mean_absolute_error
````

````python
# 데이터 전처리1 : 독립변수와 종속변수 나누기
df_x = raw_data[["RM", "RAD"]]
df_y = raw_data[["MEDV"]]

# 데이터 전처리2 : 학습(train)데이터와 테스트(test)데이터 나누기
train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=0.3, random_state=1)
# 선형회기 모델 학습
model = linear_model.LinearRegression()
model.fit(train_x, train_y)

# 모델 평가
pred_y = model.predict(test_x)
#flatten/ 1차원 정렬
pred_y = np.around(pred_y.flatten(), 2)
mae = mean_absolute_error(test_y, pred_y)
mae = round(mae, 2)

# 주택 가격 예측
def make_df(rm, rad) :
    return pd.DataFrame({ "RM" : [rm], "RAD" : [rad] })

rm, rad = 5, 3
predict = model.predict(make_df(rm, rad)).flatten()[0]
predict = round(predict, 2)
````
