### quiz
- makedata 모듈을 이용해서 데이터 프레임 만들기
- user_df 
    - 8명의 데이터가 있는 데이터 프레임을 만드세요.
    - UserID : 1~8
    - Name : makedata.get_name()
    - Age : makedata.get_age()
    - 중복되는 Name 값이 없도록

```python
# 딕셔너리의 리스트 # 내가 작성
user_name = []
user_age = []
datas = {}
while True:
    
    name = makedata.get_name()
    
    if len(user_name) < 8:
        if name not in user_name:
            user_name.append(name)
            user_age.append(makedata.get_age())
    else :
        datas = {
            "Age": user_age,
            "Name": user_name
        }
        break

user_df = pd.DataFrame(datas)
user_df["UserID"] = range(1, 9)
user_df
```

```python
# 딕셔너리의 리스트 : UserID, Name, Age : 퀴즈 해설
datas = {}
datas["UserID"] = list(range(1, 9))
datas["Age"] = [makedata.get_age() for _ in range(8)]
names = []
while True :
    name = makedata.get_name()
    if name not in names :
        names.append(name)
    if len(names) >= 8 :
        break

datas["Name"] = names

user_df = pd.DataFrame(datas)
user_df
```

```python
# 리스트의 딕셔너리 # 내가 작성

datas = []
ovl_name = []

while True:

    name = makedata.get_name()

    if len(ovl_name) < 8:
        if name not in ovl_name:
            ovl_name.append(name)  # 닉네임 선저장
            data = {"Age": makedata.get_age(), "Name": name}
            datas.append(data)
    else:
        break

user_df = pd.DataFrame(datas)
user_df["UserID"] = range(1, 9)
user_df = user_df[["UserID","Age","Name"]]
user_df
```

```python
# 리스트의 딕셔너리 : UserId, Name, Age #퀴즈 해설
# 딕셔너리 데이터를 데이터 프레임에 하나씩 추가하기
user_df = pd.DataFrame(columns=["UserID", "Name", "Age"])

datas = []

for idx in range(1, 9) :
    
    name = makedata.get_name()
    while name in list(user_df["Name"]) :
        name = makedata.get_name()
    
    data = {"Name" : name, "UserID" : idx, "Age" : makedata.get_age()}
    
    user_df.loc[len(user_df)] = data
    
# user_df = pd.DataFrame(datas)
user_df
```

### quiz
- money_df 만들기
    - 15개의 데이터
    - ID : 1 ~ 8 랜덤한 숫자 제이터
    - Money : 1000원 단위로 1000원 ~ 20000원 까지의 숫자가 저장

```python
# 딕셔너리 데이터를 데이터 프레임에 하나씩 추가하기
money_df = pd.DataFrame(columns=["ID", "Money"])
# np.random.randint()

for _ in range(15) :

    money_df.loc[len(money_df)] = {
        "ID" : np.random.randint(1, 9),
        "Money" : np.random.randint(1, 21) * 1000,
    }
    
# 컬럼데이터에서 Unique 값 확인
ids = money_df["ID"].unique()
ids.sort()
ids
```

### 1. Merge 

```python
# 데이터 병합후 중복 제거 로직
```

```python
user_df.merge(money_df, left_on="UserID", right_on="ID")
```

```python
### 컬럼명 변경
user_df.rename(columns={"UserID" : "ID"}, inplace=True)
user_df.tail(1)
```

```python
user_df.merge(money_df).tail()
```

```python
result_df = pd.merge(money_df, user_df)
result_df.tail()
```

```python
# groupby, sum, size, min ..함수 : Series.reset_index()
money_list = result_df.groupby("Name").sum()["Money"].reset_index()
money_list
```

```python
# groupby : agg("sum"), agg("mean") .. : DataFrame
money_list = result_df.groupby("Name").agg("sum").reset_index()[["Name", "Money"]]
money_list
```

```python
# merge : money_list, user_df : outer
```

```python
result = pd.merge(user_df, money_list, how="outer" )
result
```

```python
# fillna: NaN 을 특정 데이터로 채워줌
```

```python
result.fillna(value=0, inplace=True)
result
```

```python
# money 칼럼을 정수 데이터 타입을 변경
result.dtypes
```

```python
result["Money"] = result["Money"].astype("int")
result
```

```python
result.sort_values("Money", ascending=False)[:3] #sort 데이터중 상위 3
```
